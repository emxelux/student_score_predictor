```python
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
```

```python
x = df.drop("Performance Index", axis = 1)
y = df['Performance Index']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

```python
x_train_scaler = StandardScaler()
x_test_scaler = StandardScaler()
# y_train_scaler = StandardScaler()
# y_test_scaler = StandardScaler()
```

```python
x_train_scaled = x_train_scaler.fit_transform(x_train)
x_test_scaled = x_test_scaler.fit_transform(x_test)
# y_train_scaled = y_train_scaler.fit_transform(y_train)
# y_test_scaled = y_test_scaler.fit_transform(y_test)
```

```python
model = LinearRegression()
model.fit(x_train_scaled, y_train)
```

```python
y_pred = model.predict(x_test_scaled)
```

```python
train_pred = model.predict(x_train)
```

```python
print(f'train_score: {model.score(x_train_scaled, y_train)}')
print(f'test_score: {model.score(x_test_scaled, y_test)}')
```

```python
with open('reg_model.pkl', 'wb') as f:
    pk.dump(model, f)
```
